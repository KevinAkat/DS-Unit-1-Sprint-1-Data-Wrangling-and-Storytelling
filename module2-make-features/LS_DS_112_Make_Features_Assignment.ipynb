```python
# Convert the term column from string to integer

# Creating a function to convert a single string to integer

def str_to_int(x):
  return int(x.strip(' months'))

# Using the apply function to use str_to_int function on "term" Column

df_loan['term'] = df_loan['term'].apply(str_to_int)

# Checking to see if it worked

df_loan['term'].head()
```

```python
# Using Isin function to convert True/False Booleans into 1/0 Ints
# If the Status is Fully Paid

df_loan['loan_status_is_great'] = df_loan['loan_status'].isin(['Current','Fully Paid']).astype(int)

# Checking to see if it worked

df_loan['loan_status_is_great'].head(25)
```

```python
# Making last_pymnt_d_month and last_pymnt_d_year columns

# Converting "Last_Pymnt_D" from string to datetime format

df_loan['last_pymnt_d'] = pd.to_datetime(df_loan['last_pymnt_d'], 
                                          infer_datetime_format=True).head()

```

```python
# Creating the last_pymnt_d_month and last_pymnt_d_year columns

df_loan['last_pymnt_d_month'] = (df_loan['last_pymnt_d'].dt.month)
df_loan['last_pymnt_d_year'] = (df_loan['last_pymnt_d'].dt.year)
```
